```python
import cv2
import matplotlib.pyplot as plt
import dlib
print('modules loaded')
```

```python
img_cv = cv2.imread('lena_test_image.jpeg')
img = dlib.load_rgb_image('lena_test_image.jpeg')
plt.imshow(img_cv)
plt.show()
#print(type(img))
```

```python
img = dlib.load_rgb_image('lena_test_image.jpeg')
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

#detect the faces
faces = detector(img)

for face in faces:
  print(face.top(), face.left(), face.right(),face.bottom())

#lets draw the rect
cv2.rectangle(img, (face.left(),face.top()), (face.right(),face.bottom()), (255,255,0), 5)

#lets draw the landmarks
landmarks = predictor(img, face)
parts = landmarks.parts()
for part in parts:
  #overlay a mini circle on each landmark
  cv2.circle(img, (part.x, part.y), 1, (255,0,0), 2)
#make a blue dot at what we think is the 1st landmark
cv2.circle(img, (parts[0].x, parts[0].y), 5, (0,0,255), 2)

plt.imshow(img)
plt.show()
#detect landmarks
#print(type(faces))
#print(faces)
#dir(faces)
```

Points 49, 55 - 61, and 65 are the most useful to use in determining if someone is smiling or not.
