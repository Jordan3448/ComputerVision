```python
#grab image off the internet and display it
import cv2
import numpy as np
import matplotlib.pyplot as p

#!wget 'https://starsunfolded.com/wp-content/uploads/2016/07/Michael-Jackson.jpg'
!ls
!mv 'Michael-Jackson.jpg' mj.jpg
```

```python
image = cv2.imread('manyfaces.jpg')
img_correct_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
p.imshow(img_correct_color)
p.show()
```

```python
#generate a copy of the image
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
p.imshow(img_gray)
p.show()
```

```python
#initalize the haar wavelet CascadeClassifier
face_cascade_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
face_output = face_cascade_classifier.detectMultiScale(img_gray, 1.1, 4)
# detectMultiScale returns a set of (x, y, height, width) of face candidates
print(face_output.shape)
print('number of faces detected: ', face_output.shape[0])
```

```python
# we use opencv to draw rectanges for each face onto the original image
for face in face_output:
  x,y,height, width = face
  #plot a rectangle on original image
  cv2.rectangle(image, 
                (x,y), 
                (x+width, y+height), 
                color=(255,50,255), 
                thickness=2)
img_fixed = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
p.imshow(img_fixed)
p.show()
```
