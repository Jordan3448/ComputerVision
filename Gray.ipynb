```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
print('modules loaded')
```

a.png is a screenshot uploaded into the collab through draging and dropping into the lefthand files menu.

```python
x = cv2.imread ('a.png')
x = cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
x0 = x.copy()

print('shape of x', x.shape)
for yi in range(x.shape[0]):
  for xi in range(x.shape[1]):
    for ci in range(x.shape[2]):
#     if x[yi,xi,ci] >= 128:
#       x[yi,xi,ci] = int(x[yi,xi,ci] /2)
      if x[yi,xi,ci] >= 128:
        x[yi,xi,ci] = 255
      else: 
        x[yi,xi,ci] = 0

#cv2.imwrite('fname.jpg', x)
plt.imshow(x)
plt.show()
```

```python
print('shape of x', x.shape)
for yi in range(x.shape[0]):
  for xi in range(x.shape[1]):
    for ci in range(x.shape[2]):
      #invert y
      x[yi,xi,ci] = x0[x.shape[0]-1-yi, xi, ci]
plt.figure()
plt.imshow(x0)
plt.figure()
plt.imshow(x)
plt.show()
```

```python
print('shape of x', x.shape)
rs = 20
for yi in range(x.shape[0]):
  for xi in range(x.shape[1]-rs):
    for ci in range(x.shape[2]):
      #move right 20px
      x[yi,xi,ci] = x0[yi, xi+rs, ci]
plt.figure()
plt.imshow(x0)
plt.figure()
plt.imshow(x)
plt.show()
```
